```python
from datetime import date as dt
import pandas as pd
import numpy as np
import os
import glob
import matplotlib.pyplot as plt
import seaborn as sns
import pickle
import os.path
import io
```

# Loading in data

```python
ucla_data = pd.read_csv("ucla.csv")
usc_data = pd.read_csv('usc.csv')
ucla_and_usc = pd.read_csv('ucla_and_usc.csv')
```

```python
# Setting a large size for future plots
sns.set(rc = {'figure.figsize':(12,10)})
```

```python
wins = ucla_data['Overall W'].values
ucla_linegraph = sns.lineplot(ucla_data['Year'].values, wins,  color= 'blue').set_title("UCLA Football Wins Over Time",
                                                                                   fontsize=30)
```

```python
wins = usc_data['Overall W'].values
usc_linegraph = sns.lineplot(usc_data['Year'].values, wins,  color= 'red').set_title("USC Football Wins Over Time",
                                                                                   fontsize=30)
```

# Combining UCLA & USC

```python
# Setting a large size for future plots (width x height)
sns.set(rc = {'figure.figsize':(13,10)})

usc_vs_ucla_wins = sns.lineplot(
    'Year', 'Overall W', data = ucla_and_usc, 
    hue='Team', palette = ['blue', 'red']
    ).set_title("UCLA vs USC Football Wins Over Time", fontsize=30
)

plt.ylabel('Wins')
plt.xlim(2000, 2021)
```

# UCLA Wins Over Time

```python
# UCLA graph with horizontal line at 10 games
ucla_wins = sns.lineplot(
    'Year', 'Overall W', data = ucla_data, 
    palette = ['blue']
    ).set_title("UCLA Football Wins Since 1972", fontsize=30
)

plt.ylabel('Wins')
plt.xlim(1972, 2023)
plt.axhline(y=10, color='red', linestyle='--')
```
